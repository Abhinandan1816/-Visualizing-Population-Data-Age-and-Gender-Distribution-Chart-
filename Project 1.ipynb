## Create a bar chart or histogram to visualize the distribution of categorical or continuous variable, such as the distribution of ages or genders in a population.



Sample Dataset :- https://data.worldbank.org/indicator/SP.POP.TOTL

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("C:/Users/abhin/Downloads/1c089db0-6a4d-4d57-9827-0b0df1ceea9f_Data.csv")
```

```python
print(df.head())
```

```python
population_2022 = df[["Country Name", "2022 [YR2022]"]]
population_2022 = population_2022.dropna()
population_2022["2022 [YR2022]"] = pd.to_numeric(population_2022["2022 [YR2022]"], errors="coerce")

top_20_population = population_2022.sort_values(by="2022 [YR2022]", ascending=False).head(20)


plt.figure(figsize=(12, 6))
plt.barh(top_20_population["Country Name"], top_20_population["2022 [YR2022]"], color="skyblue")
plt.xlabel("Population (2022)")
plt.ylabel("Country")
plt.title("Top 20 Countries by Population in 2022")
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.show()
```
